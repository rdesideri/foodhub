```python
from pyspark.sql import functions as F
from pyspark.sql import types as T
```

```python
file_path = "/Workspace/Users/rdesideri@gmail.com/foodhub_order/foodhub_order.csv"
df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")
         .csv(file_path)
)

display(df)

df.printSchema()
print("Total de linhas:", df.count())

```

```python
# 1) Seleção + cast explícito (Silver = schema controlado)
df_silver = (
    df
    .select(
        F.col("order_id").cast(T.IntegerType()).alias("order_id"),
        F.col("customer_id").cast(T.IntegerType()).alias("customer_id"),
        F.col("restaurant_name").cast(T.StringType()).alias("restaurant_name"),
        F.col("cuisine_type").cast(T.StringType()).alias("cuisine_type"),
        F.col("cost_of_the_order").cast(T.DoubleType()).alias("cost_of_the_order"),
        F.col("day_of_the_week").cast(T.StringType()).alias("day_of_the_week"),
        F.col("rating").cast(T.StringType()).alias("rating"),
        F.col("food_preparation_time").cast(T.IntegerType()).alias("food_preparation_time"),
        F.col("delivery_time").cast(T.IntegerType()).alias("delivery_time"),
    )
)

df_silver.printSchema()
display(df_silver.limit(10))
```

```python
null_report = (
    df_silver.select([
        F.sum(F.when(F.col(c).isNull(), 1).otherwise(0)).alias(c)
        for c in df_silver.columns
    ])
)

display(null_report)
```

```python
dup_order_id = (
    df_silver
    .groupBy("order_id")
    .count()
    .filter(F.col("count") > 1)
)

display(dup_order_id)
```

```python
print("Quantidade de order_id duplicados:", dup_order_id.count())
```
